```python
df = pd.read_excel("Customer Call List (2).xlsx")
```

```python
df
```

```python
df = df.drop(columns = "Not_Useful_Column")
```

```python
df['Phone_Number'] = df['Phone_Number'].astype(str).str.replace(r'[^0-9a-zA-z]', '',regex = True)
```

```python
df
```

```python
df['Paying Customer'] = df['Paying Customer'].str.replace('Yes','Y').str.replace('No','N')
```

```python
df['Do_Not_Contact'] = df['Do_Not_Contact'].str.replace('Yes','Y').str.replace('No','N')
```

```python
df['Phone_Number'] = df['Phone_Number'].apply(lambda x: f'{x[0:3]}-{x[3:6]}-{x[6:10]}' if len(x) == 10 else 'Invalid')
```

```python
df['Phone_Number'] = df['Phone_Number'].str.replace('Invalid', '')
```

```python
df[['Street_Address', 'State', 'Zip_Code']]  = df['Address'].str.split(',', n=2, expand=True)
```

```python
df.drop(columns = 'Address')
```

```python
df = df.drop_duplicates()  #this should be 1st step.
```

```python
df['Last_Name'] = df['Last_Name'].str.strip('123._/')
```

```python
df
```

```python
df = df.replace('N/a', '').fillna('')
```

```python
df
```

```python
df['Do_Not_Contact'] = df['Do_Not_Contact'].fillna('')
```

```python
df
```

```python
for x in df.index:
  if df.loc[x, 'Do_Not_Contact'] == 'Y':
    df.drop(x, inplace = True)
```

```python
df = df.drop(columns = 'Do_Not_Contact')
```

```python
for x in df.index:
  if df.loc[x, 'Phone_Number'] == '':
    df.drop(x, inplace = True)
```

```python
df
```

```python
df = df.drop(columns = 'Address')
```

```python
df = df.reset_index(drop = True)
```

```python
df #cleaned data
```

```python
df.concat(df['First_Name'], df['Last_Name'])
```

```python
df['Full_Name'] = df['First_Name'] + ' ' + df['Last_Name']
display(df.head())
```

```python
new_column_order = ['CustomerID', 'Full_Name', 'Phone_Number', 'Paying Customer', 'Street_Address', 'State', 'Zip_Code']
df = df[new_column_order]
display(df.head())
```

```python
df
```

```python
df = df.drop(columns = 'Zip_Code')  #all the entries are empty, so we can drop the column
```

```python
df #final cleaned data
```

```python
df.to_excel("cleaned_customer_call_list.xlsx", index=False)
```
